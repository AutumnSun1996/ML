```python
import pandas as pd
```

```python
data = pd.read_csv('train.data', sep='\t')
data.head()
```

```python
data.describe()
```

```python
objs = data.describe(include=[object]).T.sort_values('unique', ascending=False)
objs
```

```python
data = data.dropna(axis=1, how='all')
```

```python
mean_val = data.mean()
indices = mean_val.index
data[indices] = (data[indices] - mean_val) / data[indices].std()
```

```python
print(data.shape)
dummy = pd.get_dummies(data).fillna(0)
print(dummy.shape)
```
